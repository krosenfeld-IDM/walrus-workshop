```python
import os
import pickle

from pathlib import Path
from dataclasses import dataclass
```

```python
@dataclass
class Feature:
    index: int
    corr: float
    p_value: float

    def __lt__(self, other):
        return self.corr < other.corr
```

```python
data_dir = Path("metrics") / "enstrophy_rho" 
all_data = pickle.load(open(data_dir / "feature_list_enstrophy.pkl", "rb"))

data_dir = Path("figures/preprint/data")
traj = {}
traj['50'] = pickle.load(open(data_dir / "enstrophy_feature_list_traj_50.pkl", "rb"))
traj['56'] = pickle.load(open(data_dir / "enstrophy_feature_list_traj_56.pkl", "rb"))
traj['3'] = pickle.load(open(data_dir / "enstrophy_feature_list_traj_3.pkl", "rb"))
```

```python
for k,v in traj.items():
    print('Trajectory', k, 'finds features', sorted(v['feature_ids'][:5]))
print('All trajectories find features', sorted(all_data['feature_ids'][:5]))
```
